```python
import numpy as np

def printElements(matrix):
    print(matrix)
    print("Размерность(m,k) = " + str(matrix.shape))
    print("a12 = " + str(matrix[0][1]))
    print("a21 = " + str(matrix[1][0]))
    print("a22 = " + str(matrix[1][1]))

def trySumMatrix(matrixA, matrixB,printMatrix=0):
    if printMatrix == 1:
        print("Матрица А")
        print(matrixA)
        print("Матрица B")
        print(matrixB)
        print("Сумма")
    if (matrixA.shape == matrixB.shape):
        return matrixA + matrixB
    else: 
        return "Нельзя складывать матрицы"

def tryMinusMatrix(matrixA, matrixB,printMatrix=0):
    if printMatrix == 1:
        print("Матрица А")
        print(matrixA)
        print("Матрица B")
        print(matrixB)
        print("Разность")
    if (matrixA.shape == matrixB.shape):
        return matrixA - matrixB
    else: 
        return "Нельзя вычитать матрицы"

def tryMultiplyAxis(axisA, axisB):
    print("Строка А")
    print(axisA)
    print("Строка B")
    print(axisB)
    print("Произвдение")
    if (axisA.shape[1] == axisB.shape[0]):
        return (np.matmul(axisA,axisB)).sum()
    else: 
        return "Нельзя умножать строки"

def tryMultiplyMatrix(matrixA, matrixB,printMatrix=0):
    if printMatrix == 1:
        print("Матрица А")
        print(matrixA)
        print("Матрица B")
        print(matrixB)
        print("Произведение")
    if (matrixA.shape[1] == matrixB.shape[0]):
        return np.matmul(matrixA,matrixB)
    else: 
        return "Нельзя умножать матрицы"
```

```python
#1


matrixA = np.array(np.asmatrix('1 -1 2 0; 0 1 1 3'))
print("Матрица А")
printElements(matrixA)

print("Матрица Б")
matrixB = np.array(np.asmatrix('1 0 -1 3 2;0 1 1 2 0; -1 2 0 3 0'))
printElements(matrixB)

print("Матрица В")
matrixC = np.array(np.asmatrix('1 1 2; 2 1 3;3 1 0; 0 -1 1'))
printElements(matrixC)
```

```python
#2
#А
print("A)")
matrixA = np.array(np.asmatrix('1 2'))
matrixB = np.array(np.asmatrix('3;0'))
print(trySumMatrix(matrixA,matrixB,1))

#Б
print("Б)")
matrixA = np.array(np.asmatrix('0 1'))
matrixB = np.array(np.asmatrix('0 2 3'))
print(trySumMatrix(matrixA,matrixB,1))

#В
print("В)")
matrixA = np.array(np.asmatrix('1 -1; 0 2'))
matrixB = np.array(np.asmatrix('2 3;4 -3'))
print(trySumMatrix(matrixA,matrixB,1))


#Г
print("Г)")
matrixA = np.array(np.asmatrix('1 2 0 1;-1 3 -2 4'))
matrixB = np.array(np.asmatrix('2 5 -1 3;3 0 -2 4'))
print(trySumMatrix(matrixA,matrixB,1))
```

```python
#3
#А
print("A)")
axisA = np.array(np.asmatrix('1 -1'))
axisB = np.array(np.asmatrix('0;2;1'))
print(tryMultiplyAxis(axisA,axisB,1))

print("A)")
axisA = np.array(np.asmatrix('1 -1 2'))
axisB = np.array(np.asmatrix('1;-2;1'))
print(tryMultiplyAxis(axisA,axisB,1))

print("A)")
axisA = np.array(np.asmatrix('1 -1'))
axisB = np.array(np.asmatrix('0 -2'))
print(axisB.shape[0])
print(tryMultiplyAxis(axisA,axisB,1))
```

```python
#4
print("A)")
matrixA = np.array(np.asmatrix('1 -1 3;2 0 1'))
matrixB = np.array(np.asmatrix('1 1;-1 2;0 -1'))
print(tryMultiplyMatrix(matrixA,matrixB,1))

print("Б)")
matrixA = np.array(np.asmatrix('1 2 1;-1 2 1; 0 1 -1'))
matrixB = np.array(np.asmatrix('1 2 1;-1 2 1; 0 1 -1'))
print(tryMultiplyMatrix(matrixA,matrixB,1))


print("В)")
matrixA = np.array(np.asmatrix('4 2 1;4 5 2;3 2 1'))
matrixB = np.array(np.asmatrix('1 0 -1;2 1 -4;-7 -2 12'))
print(tryMultiplyMatrix(matrixA,matrixB,1))

print("Г)")
matrixA = np.array(np.asmatrix('1 -1;0 1'))
matrixB = np.array(np.asmatrix('3 -2;-1 1'))
matrixC = np.array(np.asmatrix('1 2;0 1'))
matrixD = np.array(np.asmatrix('2 1;1 -1'))

result = 2 * (tryMinusMatrix(matrixA,matrixB)) + tryMultiplyMatrix(matrixC,matrixD)
print(result)
```

```python
#8
print("А)")

matrixA = np.array(np.asmatrix('1 -1 2'))
matrixB = np.array(np.asmatrix('2 1 0 1;1 -1 0 -2;0 2 -1 1'))
print(tryMultiplyMatrix(matrixA,matrixB,1))

print("Б)")
matrixA = np.array(np.asmatrix('1 -1 2;0 1 1;1 0 -1'))
matrixB = np.array(np.asmatrix('1 0 -1;0 1 1;-1 2 1'))
matrixC = np.array(np.asmatrix('-1 0 1;0 1 1;2 0 1'))
matrixD = np.array(np.asmatrix('1 1 0;1 -1 0;0 1 1'))

result = tryMinusMatrix((tryMultiplyMatrix(matrixA,matrixB)),2*(tryMultiplyMatrix(matrixC,matrixD)))
print(result)

print("В)")
matrixA = np.array(np.asmatrix('1 -1;2 1;0 1'))
matrixB = np.array(np.asmatrix('1 -1 0;0 1 1'))
matrixC = np.array(np.asmatrix('1 -1 1 2; 0 1 1 0;0 0 1 -1'))
matrixD = np.array(np.asmatrix('1 -1 0;0 1 1;1 2 3;-1 0 1'))

result = trySumMatrix((tryMultiplyMatrix(matrixA,matrixB)),2*(tryMultiplyMatrix(matrixC,matrixD)))
print(result)
```
